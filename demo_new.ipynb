```python
# du doan gia co phieu
# train, test
# dataset ---> for train, for test
# for_train: nhin vao day va hocmo hinh -- predict thang tiep theo thi gia no .....
# for_test 
# dataset ---> train, khong co du lieu de ban so sanh doi chieu
```

```python
# train, test
# train truoc nam 2013
# test tu 2013
```

```python
# filter date < 2013 ----> df_train
# date >= 2013 ----> df_test

cond = df['Date'] < '2013'
df_train = df[cond]
```

```python
# df_test --> Date cua 2013
```

```python
# cond = df['Date'] >= '2013'
# df_test = df[cond]

df_test = df[df['Date'] >= '2013']
```

```python
df_train.sort_values(by = 'Date', ascending = True, inplace = True)
```

```python
df_train = df_train.reset_index(drop = True)
```

```python
df_train
```

```python
# feature engineering
# chung ta tao ra cac feature (dac trung) --> giup may co the phan biet hoc ra cai gia ma sai so it nhat
# feature enginering ---> tao ra them cac feature tu cac feture san co ---> sinh ra them cac feature
# feature engineering 
# domain knowledge, ky nang du lieu

# weekday ---> se phan anh gia cua ngay tiep
# feature_01 ----> weekday cua cot date
```

```python
df_train['feat_01'] = df_train['Date'].dt.dayofweek
```

```python
# df_train
# feat_01 la numeric, thu ngay ---> categorical
```

```python
df_train['feat_01'] = df_train['feat_01'].astype('str')
```

```python
df_train.info()
```

```python
df_train.head(10)
```

```python
# get the average 'Close' price of the most recent 5 days
Prev_5_avg = []
for index, row in df_train.iterrows():
    if index < 5:
        Prev_5_avg.append(0)
    else:
        i_5 = index - 5
        tmp = df_train.iloc[i_5:index]['Close'].mean()
        Prev_5_avg.append(tmp) 
```

```python
# df['Prev_5_avg'] = Prev_5_avg
df_train['feat_02'] = Prev_5_avg
```

```python
df_train
```

```python
# feat_03 --- Prev_30_avg
Prev_30_avg = []
for index, row in df_train.iterrows():
    if index < 30:
        Prev_30_avg.append(0)
    else:
        i_30 = index - 30
        tmp = df_train.iloc[i_30:index]['Close'].mean()
        Prev_30_avg.append(tmp) 
        
# shift ---> lead, lag trong sql ---> window function ---> run nhanh hon han cach minh dang lam
```

```python
df_train['feat_03'] = Prev_30_avg
# df_train
```

```python
# get the standard deviation of the 'Close' price of the most 5 and 30 days
Prev_5_std = []
for index, row in df_train.iterrows():
    if index < 5:
        Prev_5_std.append(0)
    else:
        i_5 = index - 5
        tmp = df_train.iloc[i_5:index]['Close'].std()
        Prev_5_std.append(tmp)
df_train['feat_04'] = Prev_5_std

Prev_30_std = []
for index, row in df_train.iterrows():
    if index < 30:
        Prev_30_std.append(0)
    else:
        i_30 = index - 30
        tmp = df_train.iloc[i_30:index]['Close'].std()
        Prev_30_std.append(tmp)
df_train['feat_05'] = Prev_30_std
```

```python
df_train
```

```python
# get the average and std of volume of the last 5 days
Prev_volume_5_std = []
Prev_volume_5_avg = []
for index, row in df_train.iterrows():
    if index < 5:
        Prev_volume_5_std.append(0)
        Prev_volume_5_avg.append(0)
    else:
        i_5 = index - 5
        
        # get std
        tmp = df_train.iloc[i_5:index]['Volume'].std()
        Prev_volume_5_std.append(tmp)
        
        # get mean
        tmp = df_train.iloc[i_5:index]['Volume'].mean()
        Prev_volume_5_avg.append(tmp)
        
df_train['feat_06'] = Prev_volume_5_std
df_train['feat_07'] = Prev_volume_5_avg

# get the average and std of volume of the last 30 days
Prev_volume_30_std = []
Prev_volume_30_avg = []
for index, row in df_train.iterrows():
    if index < 30:
        Prev_volume_30_std.append(0)
        Prev_volume_30_avg.append(0)
    else:
        i_30 = index - 30
        
        # get std
        tmp = df_train.iloc[i_30:index]['Volume'].std()
        Prev_volume_30_std.append(tmp)
        
        # get mean
        tmp = df_train.iloc[i_30:index]['Volume'].mean()
        Prev_volume_30_avg.append(tmp)
        
df_train['feat_08'] = Prev_volume_30_std
df_train['feat_09'] = Prev_volume_30_avg
```

```python
# tam hai long voi feature eningering
df_train_01 = df_train[['feat_01','feat_02','feat_03','feat_04','feat_05','feat_06','feat_07','feat_08','feat_09','Open']]
```

```python
df_train_01.head(10)
```

```python
# X_train, y_train
# X_train nhung featuer dung de huan luyen mo hinh
# y_train bien muc tieu, target variable, label
# cai nao la X_train, cai nao la y_train

X_train = df_train_01[['feat_01','feat_02','feat_03','feat_04','feat_05','feat_06','feat_07','feat_08','feat_09']]
y_train = df_train_01[['Open']]
```

```python
# train tren X_train sao cho cang khop voi y_train cang tot
# dung bat cu model gi de train cung duoc
# classsification / regression
# classification: phan loai -- bien muc tieu la categorical
# regression ---> hoi quy --- khop voi gia tri lien tuc
# 'feat_01' --> categorical variable -----> numerical variable ---> how to cahnge categorical --- numerical
# one-hot encoding
X_train = pd.get_dummies(X_train)
```

```python
X_train
```

```python
# # training a Linear Regression model
# from sklearn.linear_model import LinearRegression

# model = LinearRegression()

# # fitting step
# model.fit(X,y)

from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```
